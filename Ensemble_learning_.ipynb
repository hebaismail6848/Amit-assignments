#Decision_tree

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=.2,random_state=42)
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

#Bagging(same model)

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=load_iris()
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=.2,random_state=42)
```

```python
bag_clf=BaggingClassifier(
    base_estimator=DecisionTreeClassifier(),
    n_estimators=500,
    bootstrap=True,
    n_jobs=-1
)
```

```python
bag_clf.fit(x_train,y_train)
```

```python
y_pred=bag_clf.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

#Bagging(different model)

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris=load_iris()
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=.2,random_state=42)
```

```python
tree_clf=DecisionTreeClassifier()
logis_clf=LogisticRegression()
knn_clf=KNeighborsClassifier()
```

```python
tree_bagging=BaggingClassifier(
    base_estimator=tree_clf,
    n_estimators=500,
    bootstrap=True,
    random_state=42
)
logis_bagging=BaggingClassifier(
    base_estimator=logis_clf,
    n_estimators=500,
    bootstrap=True,
    random_state=42
)

knn_bagging=BaggingClassifier(
    base_estimator=knn_clf,
    n_estimators=500,
    bootstrap=True,
    random_state=42
)
```

```python
#tree
tree_bagging.fit(x_train,y_train)
tree_pred= tree_bagging.predict(x_test)

#logistic
logis_bagging.fit(x_train,y_train)
logis_pred= logis_bagging.predict(x_test)

#knn
knn_bagging.fit(x_train,y_train)
knn_pred= knn_bagging.predict(x_test)
```

```python
print(tree_pred)
print('**********************')
print(logis_pred)
print('**********************')
print(knn_pred)
```

```python
final_pred=np.round((tree_pred +logis_pred +knn_pred) / 3 )
final_pred
```

```python
accuracy=accuracy_score(y_test,final_pred)
accuracy
```

#AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
iris=load_iris()
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=.2,random_state=42)
```

```python
Ada_boost=AdaBoostClassifier(

    base_estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=200,
    learning_rate=.5,
    random_state=42
)
```

```python
Ada_boost.fit(x_train,y_train)
```

```python
y_pred=Ada_boost.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_pred,y_test)
print(accuracy)
```
